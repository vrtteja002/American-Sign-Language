```python
import zipfile
import os

# Define the directory to zip
directory_to_zip = '/kaggle/working/runs'

# Define the name for the zip file
zip_file_name = '/kaggle/working/output.zip'

# Create a ZipFile object in write mode
with zipfile.ZipFile(zip_file_name, 'w') as zipf:
    # Iterate over all the files and subdirectories in the directory
    for root, dirs, files in os.walk(directory_to_zip):
        for file in files:
            # Get the full path of the file
            file_path = os.path.join(root, file)
            # Add the file to the zip file
            zipf.write(file_path, os.path.relpath(file_path, directory_to_zip))

print(f'Contents of {directory_to_zip} zipped to {zip_file_name}')
```
